## 4.5 · Segment Gap Analysis *(before building the DataFrame)*

Understanding the **silence distribution between segments** informs the smoothing strategy.  
We do this *before* exploding so the decision is data-driven rather than assumed.


````python
gap_stats_train = segment_gap_stats(train_raw, n_samples=len(train_raw))
print_dict(gap_stats_train, "Inter-segment gap statistics (train, full dataset)")
gap_stats_test = segment_gap_stats(test_raw, n_samples=len(test_raw))
print_dict(gap_stats_test, "Inter-segment gap statistics (test, full dataset)")
````

````python
n_gap_samples = len(train_raw)
# Pass smoothing thresholds from config so printed labels stay in sync with audit_config.yaml
visualize_gap_distribution(
    train_raw, n_gap_samples, FIGURES_PATH,
    zoom_cap  = audit_cfg["audio"]["gap_zoom_cap"],
    small_gap = audit_cfg["audio"]["smoothing"]["small_gap"],
    large_gap = audit_cfg["audio"]["smoothing"]["large_gap"],
)
````

> **Remark — smoothing is optional but informed by this distribution.**  
> - Gaps **< 200 ms** → snap (`end = next.start`) to avoid clipping phonemes.  
> - Gaps **> 1 s**    → genuine silence; small extension (≤ 300 ms) can help the model learn end-of-utterance acoustics.  
> - Whether to apply smoothing should be decided at **the preprocessing phase**.  
> - The `smoothing_applied` boolean column in the segment DataFrame tracks this state.


## 5 · `segments` Dissection — Flat DataFrame

### Why we must explode first

The `segments` column stores a **list of dicts** per recording:
```python
[
  {"start": 0.0, "end": 3.4, "transcript": "كيفاش", "transcript_raw": "كيفاش"},
  {"start": 3.9, "end": 7.1, "transcript": "قلتلك",  "transcript_raw": "قلتلك"},
  ...
]
```
Analyzing at recording level gives us:
- One blended transcript (may mask per-segment quality issues)
- One coarse duration (last `end` time, not actual speech time)

Exploding gives us the **true unit of ASR training**: one row = one segment = one `(audio_chunk, text)` pair.

### Exhaustive schema

| Column | Type | Description |
|--------|------|-------------|
| `audio_id` | str | `"{rec_idx}_{seg_idx}"` |
| `seg_start` | float | segment start (s) |
| `seg_end` | float | segment end (s) |
| `transcript` | str | cleaned transcript |
| `transcript_raw` | str | raw transcript |

### 6.8 Code-switching (Arabic ↔ Latin) — stats + pie chart

````python
cs_seg_train  = code_switch_ratio(train_texts)
cs_seg_test   = code_switch_ratio(test_texts)
cs_word_train = code_switch_word_ratio(train_texts)
cs_word_test  = code_switch_word_ratio(test_texts)

print(f"{'Metric':<45}  {'Train':>8}  {'Test':>8}")
print("-" * 65)
print(f"{'Segments with ≥1 Latin char (has_latin %)':<45}  "
      f"{cs_seg_train:>8.2%}  {cs_seg_test:>8.2%}")
print(f"{'Words containing Latin chars':<45}  "
      f"{cs_word_train:>8.2%}  {cs_word_test:>8.2%}")
print()
print(f"Arabic words % — train: {train_text_stats['arabic_words_pct']:.2f}%  "
      f"test: {test_text_stats['arabic_words_pct']:.2f}%")
print(f"Latin  words % — train: {train_text_stats['latin_words_pct']:.2f}%  "
      f"test: {test_text_stats['latin_words_pct']:.2f}%")
````

````python
def word_type_counts(texts):
    ar, lat, other = 0, 0, 0
    for t in texts:
        for w in t.split():
            if ARABIC_CHAR_RANGE.search(w):
                ar  += 1
            elif LATIN_LETTERS.issuperset(set(w.lower().replace(" ", ""))):
                lat += 1
            else:
                other += 1
    return ar, lat, other

ar_tr, lat_tr, oth_tr = word_type_counts(train_texts)
ar_te, lat_te, oth_te = word_type_counts(test_texts)

visualize_word_pie_chart(train_texts, test_texts,
                         save_path=f"{FIGURES_PATH}/code_switch_pie.png")
````

## 8 · Cross-Analysis — Speech Rate & Audio-Text Correlation

````python
df_train["speech_rate"] = np.where(
    df_train["seg_duration"] < 1,
    df_train["word_count"],
    df_train["word_count"] / df_train["seg_duration"].replace(0, np.nan)
)
df_test["speech_rate"] = np.where(
    df_test["seg_duration"] < 1,
    df_test["word_count"],
    df_test["word_count"] / df_test["seg_duration"].replace(0, np.nan)
)

print("Speech rate (words/s) — train:")
print(df_train["speech_rate"].describe().round(3))
print()
print("Speech rate (words/s) — test:")
print(df_test["speech_rate"].describe().round(3))
````
